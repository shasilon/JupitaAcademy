##### Clean up Amount features

```python
df = df.rename(columns={'Release Clause': 'Release_clause'})
```

```python
def convert_amount(Value):
    multiplier = 1
    Value = Value.replace('€', '')
    
    if 'K' in Value:
        multiplier = 1000
        Value = Value.replace('K', '')
    elif 'M' in Value:
        multiplier = 1000000
        Value = Value.replace('M', '')
        
    amount_numeric = float(Value) * multiplier
    return '€{:,.0f}'.format(amount_numeric)

df['Value'] = df['Value'].apply(convert_amount)
```

```python
def convert_amount(Wage):
    multiplier = 1
    Wage = Wage.replace('€', '')
    
    if 'K' in Wage:
        multiplier = 1000
        Wage = Wage.replace('K', '')
    elif 'M' in Wage:
        multiplier = 1000000
        Wage = Wage.replace('M', '')
        
    amount_numeric = float(Wage) * multiplier
    return '€{:,.0f}'.format(amount_numeric)

df['Wage'] = df['Wage'].apply(convert_amount)
```

```python
def convert_amount(Release_clause):
    multiplier = 1
    Release_clause = Release_clause.replace('€', '')
    
    if 'K' in Release_clause:
        multiplier = 1000
        Release_clause = Release_clause.replace('K', '')
    elif 'M' in Release_clause:
        multiplier = 1000000
        Release_clause = Release_clause.replace('M', '')
        
    amount_numeric = float(Release_clause) * multiplier
    return '€{:,.0f}'.format(amount_numeric)

df['Release_clause'] = df['Release_clause'].apply(convert_amount)
```
