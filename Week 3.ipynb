### DATA VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

##### *Create a line plot*

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]

plt.plot(x, y, linestyle = '-', color = 'red', marker = 'o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.show()
```

#### *Create a bar plot*

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

plt.figure(figsize = (6,4))
sns.barplot(x = categories, y = values, palette = 'Set1', saturation = 1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.tight_layout
plt.show()
```

##### *Create a histogram*

```python
data =  [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]

sns.histplot(data, bins = 10, kde = False, color = 'yellow')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.tight_layout
plt.show()
```

#### *Create a scatter plot*

```python
x = [3, 5, 7, 10, 12, 15]
y = [25, 40, 30, 50, 35, 60]

sns.scatterplot(x = x, y = y, color = 'orange', marker = '*', s = 150)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.show()
```

#### *Create a pie chart*

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 15]

plt.pie(sizes, labels = labels, autopct = '%1.0f%%', startangle = 90, counterclock = False)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

#### *Create a box plot*

```python
points_data = [45, 60, 75, 300, 80, 350, 90, 95, 100, 110, 120, 250, 200]

sns.boxplot(x = points_data, linewidth = 1.5, palette = 'Set2')
plt.xlabel('Points Data')
plt.title('Box Plot')
plt.show()
```

#### *Create a barh plot*

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

plt.barh(categories, values, color = 'gold', edgecolor = 'red')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Barh Plot')
plt.show()
```
